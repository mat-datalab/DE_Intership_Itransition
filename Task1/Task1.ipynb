```python
import json

EUR_TO_USD = 1.1525

def parse_price(price_str):
    s = price_str.strip()
    symbol = s[0]
    amount = float(s[1:])
    return amount, symbol

with open("pretty_task1.json", "r", encoding="utf-8") as f:
    data = json.load(f)

for obj in data:
    price = obj.get("price")
    if not price:
        obj["USD Price"] = None
        continue

    amount, symbol = parse_price(price)

    if symbol == "€":
        usd_amount = amount * EUR_TO_USD
    else:
        usd_amount = amount

    obj["USD Price"] = round(usd_amount, 2) if usd_amount is not None else None

with open("pretty_task1.json", "w", encoding="utf-8") as f:
    json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
import json
import csv

# 1. Wczytujemy JSON
with open("pretty_task1.json", "r", encoding="utf-8") as f:
    data = json.load(f)   # data = lista słowników

# 2. Tworzymy plik CSV
with open("books.csv", "w", encoding="utf-8", newline="") as f:
    fieldnames = ["id", "title", "author", "genre", "publisher", "year", "price", "price_usd"]
    writer = csv.DictWriter(f, fieldnames=fieldnames)

    # nagłówek (HEADER) – przyda się do importu w pgAdmin
    writer.writeheader()

    for book in data:
        writer.writerow({
            "id":        book.get("id"),
            "title":     book.get("title"),
            "author":    book.get("author"),
            "genre":     book.get("genre"),
            "publisher": book.get("publisher"),
            "year":      book.get("year"),
            "price":     book.get("price"),
            # w JSON masz pewnie "USD Price", mapujemy na price_usd:
            "price_usd": book.get("USD Price")
        })
```
