```python
import re
import json

with open(r"C:\Users\Dell\Desktop\NaukaStazDE\Zadania\Task_1\task1_d.json", "r") as f:
    txt = f.read()

txt = re.sub(r':(\w+)=>', r'"\1":', txt)

data = json.loads(txt)

with open("pretty_task1.json", "w") as f:
    data = json.dump(data, f, indent=4, ensure_ascii=False)
```

```python
import json
import csv

with open("pretty_task1.json", "r", encoding="utf-8") as f:
    data = json.load(f)

with open("books.csv", "w", encoding="utf-8", newline="") as f:
    writer = csv.writer(f)
    writer.writerow(["id", "title", "author", "genre", "publisher", "year", "price"])
    
    for book in data:
        writer.writerow([
            book.get("id"),
            book.get("title"),
            book.get("author"),
            book.get("genre"),
            book.get("publisher"),
            book.get("year"),
            book.get("price")
        ])
```
